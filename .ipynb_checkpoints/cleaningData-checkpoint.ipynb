```python
import pandas as pd
import re
```

```python
# import data
path = "./comunity_developer_facebook.csv"
df = pd.read_csv(path)
df.head()
```

```python
# created new dataframe
new_df = pd.DataFrame()
```

```python
# cleaning name Author
def cleaning_author(text) :
    if 'ได้แชร์ลิงก์' in text :
        text = text.replace('ได้แชร์ลิงก์','')
    if 'ได้แชร์งานกิจกรรม' in text :
        text = text.replace('ได้แชร์งานกิจกรรม', '')
    if 'ได้แชร์โพสต์' in text :
        text = text.replace('ได้แชร์โพสต์', '')
    return text.strip()

new_df['author'] = df['author_action'].apply(cleaning_author)
```

```python
# get action
def get_action(text) :
    '''
        get action of each box.
    '''
    if "แชร์โพสต์" in text :
        return "share_post"
    if "แชร์ลิงก์" in text :
        return "share_link"
    if "แชร์งานกิจกรรม" in text :
        return "share_activities"
    else :
        return "post"
new_df['Action'] = df['author_action'].apply(get_action)
```

```python
# get like, comment, share
def get_collaborate(text) :
    ob = text.strip().split('\n')
    cnt_like = 0
    cnt_comment = 0 
    cnt_share = 0
    if len(ob) != 0 :
        try :
            cnt_like = int(ob[0])
        except :
            pass

    for idx in range(1,len(ob)) :
        if 'ความคิดเห็น' in ob[idx]:
            com = ob[idx].split()
            try :
                if com[0] == 'ความคิดเห็น' :
                    cnt_comment =  int(com[1])
                else :
                    cnt_comment =  int(com[0])
            except :
                pass
                
        if 'แชร์' in ob[idx] :
            com = ob[idx].split()
            try :
                if com[0] == 'แชร์' :
                    cnt_share =  int(com[1])
                else :
                    cnt_comment =  int(com[0])
            except :
                pass
                
    return (cnt_like, cnt_comment, cnt_share)
```

```python
collect = df['collaborate'].apply(get_collaborate)
new_df['like'] = [k[0] for k in collect]
new_df['comment'] = [k[1] for k in collect]
new_df['share'] = [k[2] for k in collect]
```

```python
# copy content
new_df['content'] = df['content']
```
